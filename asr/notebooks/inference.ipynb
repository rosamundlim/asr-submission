This notebook explores experimentation with inferencing Wav2Vec with guide from the following websites: <br>
i. https://huggingface.co/docs/transformers/en/tasks/asr#inference <br>

Import Libraries

```python
from IPython.display import Audio
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch
import pandas as pd
import librosa
```

Experimenting with single audiofile from ../cv-valid-dev/

Load sample recording

```python
sample_audio_filepath = "../cv-valid-dev/sample-000000.mp3"
y, sample_rate = librosa.load(sample_audio_filepath)
Audio(data=y, rate=sample_rate)
```

Print duration of recording

```python
print(librosa.get_duration(y=y, sr=sample_rate), "secs")
```

Resample the frequency to 16000

```python
sampling_frequency = 16000
y_16 = librosa.resample(y, orig_sr=sample_rate, target_sr=sampling_frequency)
```

Instantiate models

```python
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h")
```

Tokenize

```python
input_values = processor(y_16, sampling_rate=sampling_frequency, return_tensors="pt", padding="longest").input_values  # Batch size 1
```

Retrieve Logits, Take argmax and decode

```python
logits = model(input_values).logits
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)
```

Inspect Transcription

```python
print(transcription[0])
```
